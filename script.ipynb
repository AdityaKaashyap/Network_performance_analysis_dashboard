```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_json("telecom_tower_usaged.json")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="operator", y="dropped_calls", estimator="mean", ci=None, palette="viridis")
plt.title("Average Dropped Calls per Operator")
plt.ylabel("Dropped Calls")
plt.xlabel("Operator")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["latency_sec"], bins=20, kde=True, color="steelblue")
plt.title("Latency Distribution Across Towers")
plt.xlabel("Latency (sec)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="tower_load_percent", y="dropped_calls", hue="operator", alpha=0.7)
plt.title("Tower Load vs Dropped Calls")
plt.xlabel("Tower Load (%)")
plt.ylabel("Dropped Calls")
plt.legend(title="Operator")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="network_type", y="download_speed_mbps", palette="Set2")
plt.title("Download Speeds by Network Type")
plt.ylabel("Download Speed (Mbps)")
plt.xlabel("Network Type")
plt.show()
```

```python
import folium

# Initialize map centered on UK
m = folium.Map(location=[54.0, -2.0], zoom_start=6)

# Add towers
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row["location_latitude"], row["location_longitude"]],
        radius=5,
        popup=f"Tower {row['tower_id']}<br>Operator: {row['operator']}<br>Dropped Calls: {row['dropped_calls']}",
        color="red" if row["dropped_calls"] > 5 else "green",
        fill=True,
        fill_opacity=0.7
    ).add_to(m)

m.save("tower_map.html")
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(12,5))
df["latency_sec"].plot()
plt.title("Network Latency Over Time")
plt.ylabel("Latency (sec)")
plt.xlabel("Time")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12,6))
sns.lineplot(data=df, x=df.index, y="dropped_calls", hue="operator")
plt.title("Dropped Calls Over Time by Operator")
plt.ylabel("Dropped Calls")
plt.xlabel("Time")
plt.show()
```

```python
# Resample to 30-minute average latency
latency_trend = df["latency_sec"].resample("30min").mean()

plt.figure(figsize=(12,5))
latency_trend.plot()
plt.title("Average Latency (30-min intervals)")
plt.ylabel("Latency (sec)")
plt.show()
```

```python
df["latency_zscore"] = (df["latency_sec"] - df["latency_sec"].mean()) / df["latency_sec"].std()

# Flag anomalies
anomalies = df[df["latency_zscore"].abs() > 2]

plt.figure(figsize=(12,5))
plt.plot(df.index, df["latency_sec"], label="Latency")
plt.scatter(anomalies.index, anomalies["latency_sec"], color="red", label="Anomalies")
plt.legend()
plt.title("Latency with Anomalies Detected")
plt.show()
```

```python
from prophet import Prophet

# Prepare data for Prophet
ts_df = df.reset_index()[["timestamp", "latency_sec"]]
ts_df.columns = ["ds", "y"]

# Train model
model = Prophet()
model.fit(ts_df)

# Forecast next 1 day (1440 minutes = 24h * 60min)
future = model.make_future_dataframe(periods=1440, freq="min")
forecast = model.predict(future)

# Plot forecast
model.plot(forecast)
plt.title("Latency Forecast")
plt.show()
```
